```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/data.csv",encoding = "latin-1")
```

```python
print("Shape",df.shape)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#dropping the missing rows in which customer id is missing
df.dropna(subset=['CustomerID'],inplace = True)
```

```python
df.isnull().sum()
```

```python
df.sample(10)
```

```python
# Check for cancelled orders
cancelled = df[df['InvoiceNo'].astype(str).str.startswith('C')]
print("Cancelled orders found:", len(cancelled))
```

```python
#remove cancelled order if any
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
```

```python
#removing the bad quantites and prices
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]
```

```python
df.describe()
```

```python
df.info()
```

```python
#fix the data type of the Invoice data because it is object
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df['CustomerID'] = df['CustomerID'].astype(int).astype(str)
```

```python
df.info()
```

```python
#revenue
df['Revenue'] = df['Quantity'] * df['UnitPrice']
```

```python
df['Year']      = df['InvoiceDate'].dt.year
df['Month']     = df['InvoiceDate'].dt.month
df['YearMonth'] = df['InvoiceDate'].dt.to_period('M').astype(str)
```

```python
df.head(5)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.to_csv('../data/clean_data.csv', index=False)
```
